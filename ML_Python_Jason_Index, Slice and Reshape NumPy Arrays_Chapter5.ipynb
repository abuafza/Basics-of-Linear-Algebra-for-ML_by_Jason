Listing 5.1: Example of creating one-dimensional array.

```
# create one-dimensional array
from numpy import array
# list of data
data = [11, 22, 33, 44, 55]
# array of data
data = array(data)
print(data)
print(type(data))
```

Listing 5.3: Example of creating two-dimensional array.

```
# create two-dimensional array
from numpy import array
# list of data
data = [[11, 22],
[33, 44],
[55, 66]]
# array of data
data = array(data)
print(data)
print(type(data))
```

Listing 5.5: Example of indexing a one-dimensional array.

```
# index a one-dimensional array
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
# index data
print(data[0])
print(data[4])
```

Listing 5.9: Example of negative indexing a one-dimensional array.

Listing 5.18: Example of slicing a one-dimensional array. data[from:to]

```
# slice a one-dimensional array
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
print(data[:])
```

Listing 5.20: Example of slicing a subset of a one-dimensional array.

```
# slice a subset of a one-dimensional array
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
print(data[0:1])
```

Listing 5.22: Example of slicing with a negative index on a one-dimensional array.

```
# negative slicing of a one-dimensional array
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
print(data[-2:])
```

Listing 5.26: Example of slicing a dataset into input and output variables.

```
# split input and output data
from numpy import array
# define array
data = array([
[11, 22, 33],
[44, 55, 66],
[77, 88, 99]])
# separate data
X= data[:, :-1]
print(X)
```

Listing 5.30: Example of slicing a dataset into train and test subsets

```
# split train and test data
from numpy import array
# define array
data = array([
[11, 22, 33],
[44, 55, 66],
[77, 88, 99]])
# separate data
split = 2
train,test = data[:split,:],data[split:,:]
print(train)
print(test)
```

Listing 5.32: Example of accessing shape for a one-dimensional array

```
# shape of one-dimensional array
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
print(data.shape)
```

Listing 5.34: Example of accessing shape for a two-dimensional array.

```
# shape of a two-dimensional array
from numpy import array
# list of data
data = [[11, 22],
[33, 44],
[55, 66]]
# array of data
data = array(data)
print(data.shape)
```

Listing 5.36: Example of accessing shape for a two-dimensional array in terms of rows and
columns.

```
# row and column shape of two-dimensional array
from numpy import array
# list of data
data = [[11, 22],
        [33, 44],
[55, 66]]
# array of data
data = array(data)
print('Rows: %d' % data.shape[0])
print('Cols: %d' % data.shape[1])
```

Listing 5.39: Example of changing the shape of a one-dimensional array with the reshape()
function.

```
# reshape 1D array to 2D
from numpy import array
# define array
data = array([11, 22, 33, 44, 55])
print(data.shape)
# reshape
data = data.reshape((data.shape[0], 1))
print(data.shape)
```

Listing 5.42: Example of changing the shape of a two-dimensional array with the reshape()
function.

```
# reshape 2D array to 3D
from numpy import array
# list of data
data = [[11, 22],
[33, 44],
[55, 66]]
# array of data
data = array(data)
print(data.shape)
# reshape
data = data.reshape((data.shape[0], data.shape[1], 1))
print(data.shape)
```
